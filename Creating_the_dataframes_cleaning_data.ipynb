# Cheking format of neighborhoods

The Mariano Matamoros Centro neighborhood has not coordinates available in google or other  online, so I renamed  it as the closest neighborhood, in this case Mariano Matamoros. 

I couldn't find the coordinates of the Tres de octubre neighborhood, the problem was the "Tres", so I changed it to the number 3.

The Obrera seccion 1 and Fraccionamiento Natura neighborhoods had the same problem as Mariano Matamoros Centro, the same process was done, in this case the closest neighborhood is Obrera and Fraccionamiento Hacienda las Delicias.

The division between Sanchez Taboada Produsta and Sanchez Taboada neighborhoods is non-existent this 2 are always taken as one, so I did the same. 

```python
df3['COLONIA DEL HECHO'] = df3['COLONIA DEL HECHO'].replace({'MARIANO MATAMOROS CENTRO':'MARIANO MATAMOROS'})
df3['COLONIA DEL HECHO'] = df3['COLONIA DEL HECHO'].replace({'TRES DE OCTUBRE':'3 DE OCTUBRE'})
df3['COLONIA DEL HECHO'] = df3['COLONIA DEL HECHO'].replace({'OBRERA SECC I':'OBRERA'})
df3['COLONIA DEL HECHO'] = df3['COLONIA DEL HECHO'].replace({'OBRERA II SECCION':'OBRERA'})
df3['COLONIA DEL HECHO'] = df3['COLONIA DEL HECHO'].replace({'FRACC NATURA':'FRACC HCDA LAS DELICIAS'})
df3['COLONIA DEL HECHO'] = df3['COLONIA DEL HECHO'].replace({'SANCHEZ TABOADA PRODUTSA':'SANCHEZ TABOADA'})
df3['CLASIFICACION DEL DELITO'] = df3['CLASIFICACION DEL DELITO'].replace({'ROBO DE VEHÍCULO':'ROBO DE VEHICULO'})
df3['CLASIFICACION DEL DELITO'] = df3['CLASIFICACION DEL DELITO'].replace({'ROBO DE VEHÍCULO CON VIOLENCIA':'ROBO DE VEHICULO CON VIOLENCIA'})
df3['DELITO'] = df3['DELITO'].replace({'ROBO DE VEHÍCULO':'ROBO DE VEHICULO'})
```

Erasing the white space in the cells

```python
df3['CLASIFICACION DEL DELITO'] = df3['CLASIFICACION DEL DELITO'].map(lambda x: x.strip())
```

```python
df3['COLONIA DEL HECHO'] = df3['COLONIA DEL HECHO'].map(lambda x: x.strip())
```

```python
df3['DELITO'] = df3['DELITO'].map(lambda x: x.strip())
```

Looking for the coordinates of each neigborhood

```python
# 1 - conveneint function to delay between geocoding calls
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
# 2- - create location column
NEIGHBORHOOD_MORE_VIOLENCE['location'] =  NEIGHBORHOOD_MORE_VIOLENCE['ADDRESS'].apply(geocode)
# 3 - create longitude, laatitude and altitude from location column (returns tuple)
NEIGHBORHOOD_MORE_VIOLENCE['point'] =  NEIGHBORHOOD_MORE_VIOLENCE['location'].apply(lambda loc: tuple(loc.point) if loc else None)
# 4 - split point column into latitude, longitude and altitude columns
NEIGHBORHOOD_MORE_VIOLENCE[['latitude', 'longitude', 'altitude']] = pd.DataFrame(NEIGHBORHOOD_MORE_VIOLENCE['point'].tolist(), index= NEIGHBORHOOD_MORE_VIOLENCE.index)
```
